#  케이프 투자 증권

```python
df_Kafe = df[df['증권사'] == '케이프투자증권']
```

```python
df_Kafe['내용']
```

```python
Kafe_report = list(df_Kafe['내용'])
```

```python
len(Kafe_report)
```

```python
K = []
for kafe in Kafe_report:
    ka = re.sub('[0-9](.)[0-9]', '', kafe)
    text = re.compile('[^ ㄱ-ㅣ가-힣.]+').sub('', ka)
    text = (' ').join(re.findall('[가-힣.]+', text))

    processing = re.sub('표' , '', text)
    processing = re.sub('자료', '', processing)
    processing = re.sub('케이프투자증권', '', processing)
    processing = re.sub('그림', '', processing)
    processing = re.sub('리서치센터', '', processing)
    processing = re.sub('최운선', '', processing)
    processing = re.sub('기준시점 경제전망보고서월 본 조사는 고객의 투자에 정보를 제공할 목적으로 작성되었으며 어떠한 경우에도 무단 복제 및 배포 될 수 없습니다', '', processing)
    processing = re.sub('또한 본 에 수록된 내용은 당사가 신뢰할 만한  및 정보로 얻어진 것이나 그 정확성이나 완전성을 보장할 수 없으므로 투자자 자신의 판단과 책임하에 최종결정을 하시기 바랍니다', '', processing)
    processing = re.sub('따라서 어떠한 경우에도 본 는 고객의 주식투자의 결과에 대한 법적 책임소재의 증빙로 사용될 수 없습니다', '', processing)
    processing = processing.split('.')

    processed = []
    for pro in processing:
        pro = pro.strip()
        if pro:
            processed.append(pro)
    K.append(processed)

len(K)
```
