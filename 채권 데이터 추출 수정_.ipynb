```python
brand_list2 = []
title_list2 = []
url_list2 = []
leng_list = []
for page in range(104):
    url = 'https://finance.naver.com/research/debenture_list.nhn?'
    query = {'searchType':'writeDate',
             'writeFromDate':'2005-01-01',
             'writeToDate':'2017-12-31',
             'page': page + 1}
    
    resp = requests.get(url, query)
    soup = BeautifulSoup(resp.content, 'lxml')

    tr = soup.select('tr')
    tr_ = [a.text for a in tr]
    td = tr_.copy()
    flatten = ('').join(td)
    brand = re.findall('([가-힣A-z]+)\n\n[0-9]',flatten)
    brand_list2.extend(brand)
    
    title_soup = soup.select('td[style*="padding-left:10px"]')
    title = [a.text for a in title_soup]
    title_list2.extend(title)
    
    link_soup = soup.select('tr td.file a')
    link = [a['href'] for a in link_soup]

    url_list2.extend(link)
    
    
    
    leng_list.append([len(brand), len(title), len(link)])
    
```
